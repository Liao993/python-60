### Q1:計算標準常態分配，小於1的機率有多大?

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## Q1
#  提示: 用 cdf 來算
mu = 0
sigma = 1 #標準常態分佈 mu=0 , sigma= 1
x = np.linspace(-10,10,10000)
y = stats.norm.pdf(x,mu,sigma)
plt.plot(x,y,linewidth=2)
plt.title('pdf of Norma(0,1)')
plt.show()
```

```python
ans = stats.norm.cdf(1,mu,sigma) #累積到1 = Z<1
print("P(Z<=1) = ",ans)
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
smaller1 = stats.norm.cdf(1,mu,sigma)
print("P(Z>1) = ", 1-smaller1)

smallerminus1 = stats.norm.cdf(-1,mu,sigma)
print("P(Z<=-1) = ", smallerminus1)

ans2 = (1-smaller1) + smallerminus1
print("P(Z>1) & P(Z<=-1) = ",ans2)
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
## Q3
#  提示: 和 Q1,Q2 不同，要先標準化。
munew = 2
sigmanew = 4**(1/2)
ans3 = stats.norm.cdf(6,munew,sigmanew)
print("P(Z<=6) = ",ans3)
```
