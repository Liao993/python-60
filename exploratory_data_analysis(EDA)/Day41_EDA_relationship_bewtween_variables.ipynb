## Step3:ÈÅãÁî® researchpy Â•ó‰ª∂ÔºåË®àÁÆóÂá∫ Cramer‚Äôs V ‰øÇÊï∏


```python
crosstab, res = researchpy.crosstab(data['sex'], data['insomnia'], test='chi-square')

print(res)
print("Cramer's value is",res.loc[2,'results'])

#ÈÄôÈÇäÁî®Âç°ÊñπÊ™¢ÂÆöÁç®Á´ãÊÄßÔºåÊâÄ‰ª•Êé°Áî®ÁöÑ test ÂèÉÊï∏ÁÇ∫Âç°Êñπ "test =" argument.
# Êé°Áî®ÁöÑËÆäÊï∏Âú®ÈÄôÂÄãÊ®°ÁµÑ‰∏≠ÔºåÊúÉËá™Â∑±Ê†πÊìöË≥áÊñôÈõÜ‰æÜÂà§Êñ∑ÔºåCramer's Phi if it a 2x2 table, or Cramer's V is larger than 2x2.
```

<img src='http://cumatrixfile.cupoy.com/0000017554B6B9E5000000096375706F795F72656C656173654349/marathon/1603345870389'>

```python
## ÂØ´‰∏ÄÂÄãÂâØÁ®ãÂºèÂà§Êñ∑Áõ∏ÈóúÊÄßÁöÑÂº∑Â∫¶
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

judgment_CramerV(df,res.loc[2,'results'])
```

* Eta SquaredÊèèËø∞‰∏ÄÂÄãÈõ¢Êï£ÂûãËÆäÊï∏ÂíåÈÄ£Á∫åÂûãËÆäÊï∏ÁöÑÁõ∏ÈóúÊÄß\
* Œ∑2 ÊòØÈÄèÈÅéËÆäÁï∞Êï∏ÂàÜÊûêÁöÑÊ¶ÇÂøµÔºåÊâÄÁî¢ÁîüÂá∫‰æÜÁöÑÊåáÊ®ôÔºåÂÖ∂ÂÖ¨ÂºèÂ¶Ç‰∏ãÔºö

```python
#Â¶ÇÊûúÊ≤íÊúâÂÆâË£ùÈÅéÔºåÂÖàÊää‰∏ã‰∏ÄË°åÁ®ãÂºèÁ¢ºÊâìÈñãÔºåÂÖàÂÆâË£ùÂ•ó‰ª∂ 
import pingouin as pg
```

### Step1: ÂèñÂá∫Â§±Áú†ÂíåÈ´îÈáçË≥áÊñô
###  Step2:ÈÅãÁî® pg.anova Ë®àÁÆó‰∏âÁ®ÆËÆäÁï∞Êï∏

```python
aov = pg.anova(dv='weight', between='insomnia', data=data, detailed=True)
aov
```

### Step3:ËÆäÁï∞Êï∏ÊèõÁÆóÂæóÂà∞ Eta Squared (ùúº^ùüê)

<img src='http://cumatrixfile.cupoy.com/0000017554B6B9E5000000096375706F795F72656C656173654349/marathon/1603345870404'>

```python
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
etaSq
```

```python
def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
judgment_etaSq(etaSq)
```
